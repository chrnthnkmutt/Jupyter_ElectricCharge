```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-0.7,0.7,0.012)
X,Y = np.meshgrid(x,x)
```

```python
Q1 = 24
Q2 = -10

x1= -0.25
y1= -0.05

x2 = 0.25
y2 = 0.05

Q3 = -0.2
```

```python
V1 = Q1 / np.sqrt( (X - x1)**2+(Y - y1)**2)
V2 = Q2 / np.sqrt( (X - x2)**2+(Y - y2)**2)
V=V1+V2
V[V>np.amax(V)/10] = np.amax(V)/10
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection='3d')
ax.plot_surface(X,Y ,Q3*V)
ax.set_zlim(0,np.amax(V)/10)
ax.view_init(30,70)
```
